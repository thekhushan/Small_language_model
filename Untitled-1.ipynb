```python
# Jupyter cell: count characters and show all types
from pathlib import Path
import string
from collections import Counter

path = Path("training_data.txt")
if not path.exists():
    print(f"File not found: {path.resolve()}")
else:
    text = path.read_text(encoding="utf-8")
    total_chars = len(text)
    counts = Counter(text)
    unique_concat = ''.join(sorted(counts.keys()))

    letters = ''.join(sorted({c for c in counts if c.isalpha()}))
    digits = ''.join(sorted({c for c in counts if c.isdigit()}))
    punctuation = ''.join(sorted({c for c in counts if c in string.punctuation}))
    whitespace = sorted({c for c in counts if c.isspace()})
    others = ''.join(sorted({c for c in counts if not (c.isalpha() or c.isdigit() or c.isspace() or c in string.punctuation)}))

    def show_ws(ch):
        return ch.replace('\n','\\n').replace('\t','\\t').replace(' ','<space>')

    print(f"Total characters: {total_chars}\n")
    print("Unique characters (concatenated):")
    print(unique_concat or "(none)")
    print("\nBy type:")
    print(" Letters     :", letters or "(none)")
    print(" Digits      :", digits or "(none)")
    print(" Punctuation :", punctuation or "(none)")
    print(" Whitespace  :", ' '.join(show_ws(c) for c in whitespace) or "(none)")
    print(" Others      :", others or "(none)\n")

    print("Top 10 most common characters:")
    for ch, cnt in counts.most_common(10):
        d = show_ws(ch) if ch.isspace() else ch
        print(f" {repr(d):>8} : {cnt}")
```

```python
chars=sorted(list(set(text)))
```

```python
stoi={ch:i for i,ch in enumerate(chars)}
itos={i:ch for i,ch in enumerate(chars)}
```

```python
def encode(s):
    return [stoi[c] for c in s]
def decode(l):
    return "".join([itos[i] for i in l])
```

```python
torch.manual_seed(1337)
batch_size = 4 # how many independent sequences will we process in parallel?
block_size = 8 # what is the maximum context length for predictions?

def get_batch(split):
    # generate a small batch of data of inputs x and targets y
    data = train_data if split == 'train' else val_data
    ix = torch.randint(len(data) - block_size, (batch_size,))
    x = torch.stack([data[i:i+block_size] for i in ix])
    y = torch.stack([data[i+1:i+block_size+1] for i in ix])
    return x, y

xb, yb = get_batch('train')
print('inputs:')
print(xb.shape)
print(xb)
print('targets:')
print(yb.shape)
print(yb)

print('----')

for b in range(batch_size): # batch dimension
    for t in range(block_size): # time dimension
        context = xb[b, :t+1]
        target = yb[b,t]
        print(f"when input is {context.tolist()} the target: {target}")
```

```python
import torch
import torch.nn as nn
from torch.nn import functional as F
torch.manual_seed(1337)

class BigramLanguageModel(nn.Module):

    def __init__(self, vocab_size):
        super().__init__()
        # each token directly reads off the logits for the next token from a lookup table
        self.token_embedding_table = nn.Embedding(vocab_size, vocab_size)

    def forward(self, idx, targets=None):

        # idx and targets are both (B,T) tensor of integers
        logits = self.token_embedding_table(idx) # (B,T,C)

        if targets is None:
            loss = None
        else:
            B, T, C = logits.shape
            logits = logits.view(B*T, C)
            targets = targets.view(B*T)
            loss = F.cross_entropy(logits, targets)

        return logits, loss

    def generate(self, idx, max_new_tokens):
        # idx is (B, T) array of indices in the current context
        for _ in range(max_new_tokens):
            # get the predictions
            logits, loss = self(idx)
            # focus only on the last time step
            logits = logits[:, -1, :] # becomes (B, C)
            # apply softmax to get probabilities
            probs = F.softmax(logits, dim=-1) # (B, C)
            # sample from the distribution
            idx_next = torch.multinomial(probs, num_samples=1) # (B, 1)
            # append sampled index to the running sequence
            idx = torch.cat((idx, idx_next), dim=1) # (B, T+1)
        return idx

vocab_size = len(chars)
m = BigramLanguageModel(vocab_size)
logits, loss = m(xb, yb)
print("Initial loss:", loss.item())

optimizer = torch.optim.AdamW(m.parameters(), lr=1e-3)

batch_size = 32
for steps in range(1000):
    # sample a batch of data
    xb, yb = get_batch('train')

    # evaluate the loss
    logits, loss = m(xb, yb)
    optimizer.zero_grad(set_to_none=True)
    loss.backward()
    optimizer.step()

print("Final loss:", loss.item())

print(decode(m.generate(idx = torch.zeros((1, 1), dtype=torch.long), max_new_tokens=500)[0].tolist()))
```
